# Spatial Pattern Analysis

This script provides the reference for spatial pattern analysis

```python
import xarray as xr
import cartopy
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib
import numpy as np
import cartopy.crs as ccrs
import util as util
    
%matplotlib inline
```

```python
ds=xr.open_dataset("../nc_files/f09.nc")
ds_mask=xr.open_dataset("../nc_files/f09_mask.nc")
ds=ds.assign_coords(lon=(((ds.lon + 180) % 360) - 180))
ds=ds.reindex(lon=sorted(ds.lon))
ds_mask=ds_mask.assign_coords(lon=(((ds_mask.lon + 180) % 360) - 180))
ds_mask=ds_mask.reindex(lon=sorted(ds_mask.lon))

DJF=ds.sel(season="DJF")
JJA=ds.sel(season="JJA")
DJF_mask=ds_mask.sel(season="DJF")
JJA_mask=ds_mask.sel(season="JJA")
```
